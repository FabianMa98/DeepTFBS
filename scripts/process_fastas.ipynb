```python
import pandas as pd
import numpy as np

from Bio.SeqIO.FastaIO import SimpleFastaParser
```

```python
# Get SP-reviewed fastas (reviewed as smaller dataset for testing purposes
with open("C:/Users/Fabian/Desktop/DeepTFBS/data/uniprot_sprot.fasta") as file:
    ids = []
    sequences = []
    lens = []
    for header, sequence in SimpleFastaParser(file):
        ids.append(header)
        sequences.append(sequence)
        lens.append(len(sequence))
```

```python
df = pd.DataFrame(dict(header = ids, seqs = sequences, lens = lens))
```

```python
df
```

```python
# Get transcriptional stuff
tfs = df[df["header"].str.contains("transcription")].copy()
tfs.loc[:,"marker"] = "TF"
tfs
```

```python
joined = pd.merge(df, tfs, on=['header', 'seqs', "lens"], how='left').fillna("NoTF")
joined.to_csv("C:/Users/Fabian/Desktop/TFs.tsv", sep="\t")
```
